```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat dataset contoh
data = {
    'Tanggal': pd.date_range(start='2023-01-01', periods=100, freq='D'),
    'Kategori': ['Elektronik', 'Pakaian', 'Makanan', 'Peralatan Rumah Tangga'] * 25,
    'Unit Terjual': [20, 15, 30, 10] * 25,
    'Pendapatan': [200, 150, 300, 100] * 25
}

df = pd.DataFrame(data)

# Visualisasi Tren Penjualan dari Waktu ke Waktu
plt.figure(figsize=(12, 6))
sns.lineplot(x='Tanggal', y='Pendapatan', data=df, hue='Kategori')
plt.title('Tren Penjualan Produk dari Waktu ke Waktu')
plt.xlabel('Tanggal')
plt.ylabel('Pendapatan')
plt.legend(title='Kategori')
plt.show()

```

```python
# Visualisasi Perbandingan Penjualan Antar Kategori
plt.figure(figsize=(12, 6))
sns.barplot(x='Kategori', y='Pendapatan', data=df, estimator=sum)
plt.title('Perbandingan Penjualan Antar Kategori Produk')
plt.xlabel('Kategori')
plt.ylabel('Total Pendapatan')
plt.show()
```
